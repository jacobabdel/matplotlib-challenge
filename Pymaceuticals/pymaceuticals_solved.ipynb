```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
#from scipy.stats import sem

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')
```

```python
# Data Source files
mouse_drug_data_file = "data/mouse_drug_data.csv"
clinical_trial_data_file = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_data_df = pd.read_csv(mouse_drug_data_file)
trial_data_df = pd.read_csv(clinical_trial_data_file)
```

```python
# Combine the data into a single dataset
all_data_df = pd.merge(mouse_data_df, trial_data_df, how='inner', on='Mouse ID')
```

```python
# Display the data table for preview
all_data_df.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint
tumor_vol_df = pd.DataFrame(data=all_data_df, columns=['Mouse ID', 'Drug', 'Timepoint','Tumor Volume (mm3)'])
tumor_vol = tumor_vol_df.groupby(['Drug', 'Timepoint'])
mean_tumor_vol = tumor_vol.mean()
mean_tumor_vol
```

```python
# Convert to DataFrame
mean_tumor_vol_df = pd.DataFrame(data=mean_tumor_vol).reset_index()

# Preview DataFrame
mean_tumor_vol_df
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
tumor_vol_SE = tumor_vol.sem()
# Convert to DataFrame
tumor_vol_SE_df = pd.DataFrame(data=tumor_vol_SE).reset_index()
tumor_vol_SE_df = tumor_vol_SE_df.drop(columns='Mouse ID')
# Preview DataFrame
tumor_vol_SE_df
```

```python
# Minor Data Munging to Re-Format the Data Frames
data_reform_df = pd.pivot(mean_tumor_vol_df, values='Tumor Volume (mm3)', index='Timepoint', columns='Drug')
tumor_vol_SE_df = pd.DataFrame(data=tumor_vol_SE)
# Preview that Reformatting worked
data_reform_df
```

```python
# Generate the Plot (with Error Bars)
#Capomulin, Infubinol, Ketapril, and Placebo

fig, ax = plt.subplots()

ax.errorbar(data_reform_df.index, data_reform_df['Capomulin'], yerr=tumor_vol_SE_df.loc['Capomulin', 'Tumor Volume (mm3)'],
            fmt="--o", color="r", label="Capomulin")
ax.errorbar(data_reform_df.index, data_reform_df['Infubinol'], yerr=tumor_vol_SE_df.loc['Infubinol', 'Tumor Volume (mm3)'],
            fmt="--^", color="b", label="Infubinol")
ax.errorbar(data_reform_df.index, data_reform_df['Ketapril'], yerr=tumor_vol_SE_df.loc['Ketapril', 'Tumor Volume (mm3)'],
            fmt="--s", color="g", label="Ketapril")
ax.errorbar(data_reform_df.index, data_reform_df['Placebo'], yerr=tumor_vol_SE_df.loc['Placebo', 'Tumor Volume (mm3)'],
            fmt="--D", color="black", label="Placebo")
ax.yaxis.grid()
ax.set_xlabel("Time(Days)")
ax.set_ylabel("Tumor Volume (mm3)")
ax.set_title("Tumor Response to Treatment")
plt.legend(loc="best")

plt.autoscale()

# Show the Figure
plt.show()
```

```python
# Save the Figure
plt.savefig("Treatment.png")
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
met_resp_df = pd.DataFrame(data=all_data_df, columns=['Mouse ID', 'Drug', 'Timepoint','Metastatic Sites'])
met_resp = met_resp_df.groupby(['Drug', 'Timepoint'])
mean_met_resp = met_resp.mean()

# Convert to DataFrame
mean_met_resp_df = pd.DataFrame(data=mean_met_resp).reset_index()

# Preview DataFrame
mean_met_resp_df
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
met_resp_SE = met_resp.sem()
# Convert to DataFrame
met_resp_SE_df = pd.DataFrame(data=met_resp_SE).reset_index()
met_resp_SE_df = met_resp_SE_df.drop(columns='Mouse ID')
# Preview DataFrame
met_resp_SE_df
```

```python
# Minor Data Munging to Re-Format the Data Frames
met_data_reform_df = pd.pivot(mean_met_resp_df, values='Metastatic Sites', index='Timepoint', columns='Drug')
met_resp_SE_df = pd.DataFrame(data=met_resp_SE)
# Preview that Reformatting worked
met_data_reform_df
```

```python
# Generate the Plot (with Error Bars)

fig, ax = plt.subplots()

ax.errorbar(met_data_reform_df.index, met_data_reform_df['Capomulin'], yerr=met_resp_SE_df.loc['Capomulin', 'Metastatic Sites'],
            fmt="--o", color="r", label="Capomulin")
ax.errorbar(met_data_reform_df.index, met_data_reform_df['Infubinol'], yerr=met_resp_SE_df.loc['Infubinol', 'Metastatic Sites'],
            fmt="--^", color="b", label="Infubinol")
ax.errorbar(met_data_reform_df.index, met_data_reform_df['Ketapril'], yerr=met_resp_SE_df.loc['Ketapril', 'Metastatic Sites'],
            fmt="--s", color="g", label="Ketapril")
ax.errorbar(met_data_reform_df.index, met_data_reform_df['Placebo'], yerr=met_resp_SE_df.loc['Placebo', 'Metastatic Sites'],
            fmt="--D", color="black", label="Placebo")
ax.yaxis.grid()
ax.set_xlabel("Treatment Duration (Days)")
ax.set_ylabel("Metastatic Sites")
ax.set_title("Metastatic Spread During Treatment")
plt.legend(loc="best")

plt.autoscale()

# Show the Figure
plt.show()
```

```python
# Save the Figure
plt.savefig("Spread.png")
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
mice_df = pd.DataFrame(data=all_data_df, columns=['Mouse ID', 'Drug', 'Timepoint'])
mice_count = mice_df.groupby(['Drug', 'Timepoint']).count()

# Convert to DataFrame
mice_count_df = pd.DataFrame(data=mice_count).reset_index()
mice_count_df = mice_count_df.rename(columns={'Mouse ID':'Count of Mice'})

# Preview DataFrame
mice_count_df
```

```python
# Minor Data Munging to Re-Format the Data Frames
mice_data_reform_df = pd.pivot(mice_count_df, values='Count of Mice', index='Timepoint', columns='Drug')

# Preview that Reformatting worked
mice_data_reform_df
```

```python
# Generate the Plot (Accounting for percentages)
#Capomulin, Infubinol, Ketapril, and Placebo
fig, ax = plt.subplots()

ax.plot(mice_data_reform_df.index, (mice_data_reform_df['Capomulin']/mice_data_reform_df['Capomulin'][0]),
            "--o", color="r", label="Capomulin")
ax.plot(mice_data_reform_df.index, (mice_data_reform_df['Infubinol']/mice_data_reform_df['Infubinol'][0]),
            "--^", color="b", label="Infubinol")
ax.plot(mice_data_reform_df.index, (mice_data_reform_df['Ketapril']/mice_data_reform_df['Ketapril'][0]),
            "--s", color="g", label="Ketapril")
ax.plot(mice_data_reform_df.index, (mice_data_reform_df['Placebo']/mice_data_reform_df['Placebo'][0]),
            "--D", color="black", label="Placebo")

ax.grid()
ax.set_xlabel("Time (Days)")
ax.set_ylabel("Survival Rate (%)")

ytick_vals = ax.get_yticks()
ax.set_yticklabels([int(y*100) for y in ytick_vals])

ax.set_title("Survival During Treatment")
plt.legend(loc="best")

plt.autoscale()

# Show the Figure
plt.show()
```

```python
# Save the Figure
plt.savefig("Survival.png")
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
tumor_pc_df = pd.DataFrame(data=mean_tumor_vol_df, columns=['Drug', 'Timepoint','Tumor Volume (mm3)'])

tumor_pc = tumor_pc_df.groupby(['Drug', 'Timepoint']).sum()
tumor_pc = tumor_pc.drop([5,10,15,20,25,30,35,40], level='Timepoint').pct_change()
tumor_pc = tumor_pc.drop([0], level='Timepoint').reset_index()
tumor_pc = tumor_pc.drop(columns='Timepoint')
tumor_pc = pd.DataFrame(data=tumor_pc)
tumor_pc['Tumor Volume (mm3)'] = tumor_pc['Tumor Volume (mm3)'] * 100
tumor_pc = tumor_pc.rename(columns={'Tumor Volume (mm3)':'Percent Change'})

# Display the data to confirm
tumor_pc
```

```python
needed_pc_df = tumor_pc.drop([1,4,6,7,8,9])
needed_pc_df = needed_pc_df.reset_index(drop=True)

needed_pc_df
```

```python
#graph it!
fig, ax = plt.subplots()
ax.bar(needed_pc_df['Drug'][0],needed_pc_df['Percent Change'][0], color='g', width=-1, align='edge')
ax.annotate('{:.0%}'.format(needed_pc_df['Percent Change'][0]/100), color='w', size='large',
            xy=(needed_pc_df['Drug'][0],0), ha='right', va='bottom',xytext=(-.25, -6))

ax.bar(needed_pc_df['Drug'][1],needed_pc_df['Percent Change'][1], color='r', width=-1, align='edge')
ax.annotate('{:.0%}'.format(needed_pc_df['Percent Change'][1]/100), color='w', size='large',
            xy=(needed_pc_df['Drug'][1],0), ha='right', va='bottom',xytext=(.7, 0))

ax.bar(needed_pc_df['Drug'][2],needed_pc_df['Percent Change'][2], color='r', width=-1, align='edge')
ax.annotate('{:.0%}'.format(needed_pc_df['Percent Change'][2]/100), color='w', size='large',
            xy=(needed_pc_df['Drug'][2],0), ha='right', va='bottom',xytext=(1.7, 0))

ax.bar(needed_pc_df['Drug'][3],needed_pc_df['Percent Change'][3], color='r', width=-1, align='edge')
ax.annotate('{:.0%}'.format(needed_pc_df['Percent Change'][3]/100), color='w', size='large',
            xy=(needed_pc_df['Drug'][3],0), ha='right', va='bottom',xytext=(2.7, 0))

ax.set_yticks(np.arange(-20,80,20))
ax.set_ylabel("% Tumor Volume Change")
ax.set_title("Tumor Change Over 45 Day Treatment")
ax.grid()
```
